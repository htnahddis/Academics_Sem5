```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("tree_img.jpg") 
```

```python
if img is not None:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title('Loaded Image')
    plt.axis('off')
    plt.show()
else:
    print("Error: Image not found.")
```

```python
print(f"Shape (Height, Width, Channels): {img.shape}")
print(f"Image Datatype: {img.dtype}")
```

```python
b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
cropped = img[50:200, 100:300]
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')
plt.show()
```

```python
img2 = cv2.resize(img, (img.shape[1], img.shape[0]))
```

```python
blended = cv2.addWeighted(img, 0.2, img2, 0.3, 0)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.title('Blended Image')
plt.axis('off')
plt.show()
```

```python
ret, binary1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
ret, binary2 = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY_INV)

and_img = cv2.bitwise_and(binary1, binary2)

or_img = cv2.bitwise_or(binary1, binary2)

xor_img = cv2.bitwise_xor(binary1, binary2)

not_img = cv2.bitwise_not(binary1)
```

```python
fig, axs = plt.subplots(1,4, figsize=(16,4))
titles = ['AND', 'OR', 'XOR', 'NOT']
images = [and_img, or_img, xor_img, not_img]
for i in range(4):
    axs[i].imshow(images[i], cmap='gray')
    axs[i].set_title(titles[i])
    axs[i].axis('off')
plt.show()
```
