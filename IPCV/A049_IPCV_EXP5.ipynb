```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("cameraman_grayscale.png")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

r_min, r_max = np.min(gray_img), np.max(gray_img)
stretched = ((gray_img - r_min) / (r_max - r_min) * 255).astype(np.uint8)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(2,2,1); plt.imshow(gray_img, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(2,2,2); plt.imshow(stretched, cmap='gray'); plt.title('Stretched'); plt.axis('off')
plt.subplot(2,2,3); plt.hist(img.ravel(), bins=256, range=[0,256]); plt.title('Original Histogram')
plt.subplot(2,2,4); plt.hist(stretched.ravel(), bins=256, range=[0,256]); plt.title('Stretched Histogram')
plt.tight_layout()

plt.show()
```

```python
# Apply histogram equalization
equalized = cv2.equalizeHist(gray_img)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(2,2,1); plt.imshow(gray_img, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(2,2,2); plt.imshow(equalized, cmap='gray'); plt.title('Equalized'); plt.axis('off')
plt.subplot(2,2,3); plt.hist(img.ravel(), bins=256, range=[0,256]); plt.title('Original Histogram')
plt.subplot(2,2,4); plt.hist(equalized.ravel(), bins=256, range=[0,256]); plt.title('Equalized Histogram')
plt.tight_layout()

plt.show()
```
