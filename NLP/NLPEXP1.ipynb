```python
import nltk
nltk.download('all')
```

```python
from nltk.tokenize import word_tokenize
s1 = "On a $50,000 mortgage of 30 years at 8 percent, the monthly payment would be $366.88."
tokens = word_tokenize(s1)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
flitered_tokens = [word for word in tokens if word not in stop_words]
print(flitered_tokens)
```

```python
print(stopwords.fileids())
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
words_to_stem = ['jumping', 'jumps', 'jumped', 'jumper','fascinate','']
stemmed_tokens = [stemmer.stem(word) for word in words_to_stem]
print(stemmed_tokens)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words_to_lemmatize = ['caring', 'cared', 'care','cars']
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in words_to_lemmatize]
print(lemmatized_tokens)
```
